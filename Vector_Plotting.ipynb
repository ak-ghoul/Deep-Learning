# Vector Plotting

NUmpy will be used for mathematical things & matplotlib to plot the graph.

```
import numpy as np
import matplotlib.pyplot as mp
mp.quiver(0,0,5,6)
mp.show()
```

```
mp.quiver(0,0,3,4, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
mp.quiver(0,0,3,-4, scale_units = 'xy', angles = 'xy', scale = 1, color = 'm')
mp.xlim(-5,10)
mp.ylim(-5,10)
mp.show()
```

```
def plot(a):
  colors = ['r','m','g','b','y']
  j=0
  for i in a:
    mp.quiver(i[0],i[1],i[2],i[3],scale_units = 'xy', scale = 1, angles = 'xy',color = colors[j])
    mp.xlim(-20,20)
    mp.ylim(-20,20)
    j=j+1
    if j == 4:
      j=0
```

```
a=[(2,3,4,5),(6,4,8,9),(1,6,3,5),(1,6,-4,3)]
plot(a)
```

```
a=[np.asarray((-10,-10,4,5)),np.asarray((6,7,8,9)),np.asarray((0,0,3,5)),np.asarray((0,0,-4,3))]  # for vector addition list has to be converted into arrays because '+' will concatenate in case of lists & tuples.
plot(a)
```

```
  plot([a[0],a[1],a[2],a[3],a[1]+a[3],a[2]+a[3]]) # converted into arrays using numpy for vector addition
  
```

#Projection of Vectors

Dot product and projection of one vector over another.<br />
Numpy is used

```
b = [(0,0,3,4),(0,0,4,-1)]
def nigga(b):
  col = ['r','b','g']
  j=0
  for i in b:
    mp.quiver(i[0],i[1],i[2],i[3], scale_units = 'xy', scale = 1,angles = 'xy', color = col[j])
    mp.xlim(-5,10)
    mp.ylim(-5,10)
    j=j+1
nigga(b)    
```

```
f= (np.dot(b[0],b[1])/(np.linalg.norm(b[1])**2))*np.asarray(b[1])
print(f)
```
